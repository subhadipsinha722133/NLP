```python
import pandas as pd
```

```python
a=pd.read_csv(r"D:\CODING\PYTHON\NLP\IMDB Dataset_1.csv")
a
```

```python
a=a.iloc[:10000]
```

```python
a.duplicated().sum()
```

```python
a.drop_duplicates(inplace=True)
```

```python
a
```

```python
a["sentiment"].value_counts()
```

```python
a["review"]=a["review"].str.lower()
```

```python
import re
def remove_html(text):
    p=re.compile("<.*?>")
    return p.sub(r'',text)
```

```python
a["review"]=a["review"].apply(remove_html)
```

```python
def remove_url(text):
    p=re.compile(r'http?//\S+|www\.S+')
    return p.sub(r'',text)
```

```python
a["review"]=a["review"].apply(remove_html)
```

```python
import string
pu=string.punctuation

def remove_punction(text):
    return text.translate(str.maketrans('','',pu))
```

```python
a["review"]=a["review"].apply(remove_punction)
```

```python
from nltk.corpus import stopwords

def remove_stopword(text):
    new_text=[]
    for i in text.split():
        if i in stopwords.words("english"):
            new_text.append('')
        else:
            new_text.append(i)
    x=new_text[ : ]
    new_text.clear()
    return " ".join(x)
```

```python
a["review"]=a["review"].apply(remove_stopword)
```

```python
a
```

```python
x=a.iloc[:,0:1]
y=a.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.2)
```

```python
x_train
```

```python
y_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()
x_train_bow=cv.fit_transform(x_train["review"]).toarray()
x_test_bow=cv.transform(x_test["review"]).toarray()
```

```python
from sklearn.naive_bayes import GaussianNB

gn=GaussianNB()
gn.fit(x_train_bow,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

y_pred=gn.predict(x_test_bow)

accuracy_score(y_test,y_pred)*100
```

```python
gn.score(x_test_bow,y_test)*100
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
lor=LogisticRegression()

lor.fit(x_train_bow,y_train)
lor.score(x_test_bow,y_test)*100
```
